```python
# importando as bibliotecas
import control
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np  
```

```python
# declarando a variável no domínio da frequência
s = control.tf('s')
s
```

```python
# definindo a função de transferência do sistema
kp = 20
tau = 200
sys = kp/(tau * s + 1)
sys
```

```python
# simulado a resposta do sistema para a função degrau unitário
t, y = control.step_response(sys, T=1500, T_num=1501)
```

```python
# imprimindo a resposta do sistema no domínio do tempo
plt.plot(t, y)
plt.xlabel("tempo (s)")
plt.ylabel("y")
plt.title("Resposta do Sistema para o Degrau Unitário")
plt.grid()
plt.show()
```

```python
# salvvando os dados
t = np.transpose(t)
y = np.transpose(y)
df = pd.DataFrame({'t': t, 'y':y})
df.to_csv('resposta.csv', index=False)
```
